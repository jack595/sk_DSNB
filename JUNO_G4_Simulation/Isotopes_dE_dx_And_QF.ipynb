```python
import matplotlib.pylab as plt
import numpy as np
plt.style.use("/afs/ihep.ac.cn/users/l/luoxj/Style/Paper.mplstyle")
%config InlineBackend.figure_format = 'svg'
# plt.rcParams['figure.dpi'] = 300
# plt.rcParams['savefig.dpi'] = 300
import sys
sys.path.append("/afs/ihep.ac.cn/users/l/luoxj/root_tool/python_script/")
import periodictable as pt
from LoadMultiFiles import LoadOneFileUproot
list_elements_gen = list(pt.elements)[1:20]
list_charge = range(1,len(list_elements_gen)+1)
v_name_to_gen = []
input_list = []
for i_th,element in enumerate(list_elements_gen):
    for isotopes in element.isotopes:
        v_name_to_gen.append(f"{element.name}_{isotopes}")
        input_list.append(f"{i_th+1} {isotopes} {i_th+1} 0")

dir_ion = \
{
   "name":v_name_to_gen,
    "input_list":input_list
}
```

```python
from matplotlib.colors import LogNorm
for i_fig,name_ion in enumerate(dir_ion["name"]):
    path_file = f"/afs/ihep.ac.cn/users/l/luoxj/JUNO_G4_Simulation/run/root/{name_ion}.root"
    print(path_file)
    try:
        dir_events = LoadOneFileUproot(path_file,
                               name_branch="GdLS_log")
        dir_geninfo = LoadOneFileUproot(path_file,
                               name_branch="genInfo")
    except :
        continue

    v_dE_dx_average = []
    v_Equench = []
    v_Edep = []
    v_Einit = []
    pdgID_certain = dir_events["step_pdgID"][0][0]
    for i in range(len(dir_events["evtID"])):
        index_e = dir_events["step_pdgID"][i]==pdgID_certain
        dEquench = dir_events["step_Equench"][i][index_e]
        dEdep = dir_events["step_Edep"][i][index_e]
        if sum(dEquench)==0:
            continue
        dE_dx_average = sum(dEquench*dir_events["step_Edep"][i][index_e]/dir_events["step_dx"][i][index_e])/sum(dEquench)
        v_dE_dx_average.append(dE_dx_average)
        v_Equench.append(sum(dEquench))
        v_Edep.append(np.sum(dEdep))
        v_Einit.append(dir_geninfo["E_init"][dir_events["evtID"][i]])

    v_Equench = np.array(v_Equench)
    v_Edep = np.array(v_Edep)
    v_QuenchFactor = v_Equench/v_Edep

    fig = plt.figure(i_fig,figsize=(18,5))
    ax = fig.add_subplot(131)
    im = ax.hist2d(v_QuenchFactor, v_dE_dx_average, bins=(50,50), cmap="Blues", norm=LogNorm())
    ax.set_xlabel("Quench Factor")
    ax.set_ylabel("dE/dx [ MeV/mm ]")
    ax.set_title(f"{name_ion}")
    # fig.colorbar(im)

    max_dE_dx = 30
    ax2 = fig.add_subplot(132)
    # ax2.hist2d(v_Einit, v_dE_dx_average, bins=(np.linspace(10,1000, 90),np.linspace(0, 4,90)), cmap="Blues",norm=LogNorm())
    ax2.hist2d(v_Einit, v_dE_dx_average, bins=(np.linspace(10,1000, 90),np.linspace(0, max_dE_dx ,90)), cmap="Blues",norm=LogNorm())
    ax2.set_xlabel("Beam Energy [ MeV ]")
    ax2.set_ylabel("dE/dx [ MeV/mm ]")
    ax2.set_title(f" {name_ion}")


    ax3 = fig.add_subplot(133)
    ax3.hist2d(v_Einit, v_dE_dx_average, bins=(np.linspace(10,1000, 90),np.linspace(0, max_dE_dx,90)), cmap="Blues",norm=LogNorm())
    ax3.set_xlabel("Beam Energy [ MeV ]")
    ax3.set_ylabel("dE/dx [ MeV/mm ]")
    ax3.set_title(f"{name_ion}")
```
