```python
import matplotlib.pylab as plt
import numpy as np
plt.style.use("/afs/ihep.ac.cn/users/l/luoxj/Style/Paper.mplstyle")
%config InlineBackend.figure_format = 'png'
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
import sys
# sys.path.append("/afs/ihep.ac.cn/users/l/luoxj/root_tool/python_script/")

track_txt = np.loadtxt("/afs/ihep.ac.cn/users/l/luoxj/JUNO_G4_Simulation/build/track.txt", delimiter="\t",dtype=str)
```

```python
import re
v_xyz = []
for line in track_txt:
    # x,y,z = np.array(line.replace("  ", " ").split("  "), dtype=float)
    list_xyz = re.sub(' +', ' ', line).split(" ")
    try:
        list_xyz.remove("")
    except :
        continue
    x, y, z = list_xyz
    v_xyz.append([x,y,z])
v_xyz = np.array(v_xyz,dtype=float)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
n_points = -1
# print(v_xyz[:n_points,0])
ax.scatter(v_xyz[:n_points,0], v_xyz[:n_points,1], v_xyz[:n_points,2], s=2)
```
