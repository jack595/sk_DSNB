```python
from LoadMultiFiles import LoadMultiFilesDataframe, LoadMultiROOTFilesEOS
from copy import copy
v_L_LS = [ "1.0", "5", "10"]
# v_L_LS = ["0.5", "1.0", "2", "5", "7", "10"]
v_particles = ["H_2",   "C_12",  "Ne_20"]
# v_particles = ["H_2"]
v_L_LS_float = np.array(v_L_LS, dtype=np.float)
dir_to_df = {"ion":[], "L_LS":[], "step_pdgID":[], "step_Equench":[]}
df_time = pd.DataFrame()
N_Step = 30
for i_particle, particle in enumerate(v_particles):
    for L_LS in v_L_LS:
        dir_LS = LoadMultiROOTFilesEOS(f"root://junoeos01.ihep.ac.cn//eos/juno/user/luoxj/JUNO_G4_Simulation/OneShotSimulation/root2_addEvtIDinGen_LS_{L_LS}mm/{particle}_10cm_"+"{}.root", 3001, 3010+i_particle*N_Step, name_branch="GdLS_log")

        dir_to_df["ion"] += [particle]*len( dir_LS["step_pdgID"])
        dir_to_df["L_LS"]+= [L_LS]*len( dir_LS["step_pdgID"])
        dir_to_df["step_pdgID"] += list( dir_LS["step_pdgID"])
        dir_to_df["step_Equench"] += list( dir_LS["step_Equench"] )
df_LS = pd.DataFrame.from_dict(dir_to_df)
del dir_LS
df_LS
```

```python
from GlobalFunction import tag_parentType
from GetPhysicsProperty import NameToPDGID
from tqdm import tqdm
dict_Equench_portion_diffIon_diffLS = {}
for L_LS in v_L_LS:
    # dir_to_df_Equench_portion = {"ion":[], "L_LS":[],"name_track":[], "Equench_portion":[]}
    dict_Equench_portion_diffIon = {}

    for name_ion in v_particles:
        dict_Equench_portion = {}
        pdgID_main_track = NameToPDGID(name_ion)

        for index, row in tqdm( df_LS[(df_LS["L_LS"]==L_LS) & (df_LS["ion"]==name_ion) ].iterrows() ):
            dict_event = {"pdgID":np.array(row["step_pdgID"]), "Equench":np.array(row["step_Equench"])}
            df_event = pd.DataFrame.from_dict(dict_event)
            df_sum = df_event.groupby("pdgID").sum()
            df_sum = df_sum/np.sum(df_sum["Equench"])
            for row in df_sum.iterrows():
                name_track = tag_parentType(row[0], pdgID_main_track)
                if name_track not in dict_Equench_portion:
                    dict_Equench_portion[name_track] = [float(row[1])]
                else:
                    dict_Equench_portion[name_track].append( float(row[1]) )
        dict_Equench_portion_diffIon[name_ion] = copy(dict_Equench_portion)
    dict_Equench_portion_diffIon_diffLS[L_LS] = copy(dict_Equench_portion_diffIon)
```

```python
v_colors = list( sns.color_palette("bright") )
dict_colors = {}
v_name_track = ["main track", "electron", "others", "gamma", "positron"]
for i, name_track in enumerate( v_name_track ):
    dict_colors[name_track] = v_colors[i]
# v_ls = ["-", "--", "-."]
for L_LS in v_L_LS:
    fig,axes = plt.subplots(len(v_particles),1,sharex="col",figsize=(8,9))
    dict_Equench_portion_diffIon = dict_Equench_portion_diffIon_diffLS[L_LS]

    for i_particle, particle in enumerate( v_particles ):
        dict_Equench_portion = dict_Equench_portion_diffIon[particle]
        for i, pdgID in enumerate( dict_Equench_portion.keys() ):
            axes[i_particle].hist( dict_Equench_portion[pdgID], bins=np.linspace(0, 1, 100), histtype="step", label=pdgID, color=dict_colors[pdgID])
        axes[i_particle].semilogy()
        axes[i_particle].set_title(particle)
        axes[i_particle].set_ylabel("$N_{Events}$")
    axes[0].legend(bbox_to_anchor=(1.0, 1.0))
    axes[-1].set_xlabel("$E_{quench}^{\#}/E_{quench}^{total}$")
    plt.suptitle("$L_{LS}=$"+f"{L_LS} mm", fontsize=25)
```

```python
dict_to_df_others = {"L_LS":[], "ion":[], "portion(others)":[]}
for L_LS in v_L_LS:
    for particle in v_particles:
        dict_portion = dict_Equench_portion_diffIon_diffLS[L_LS][particle]

        dict_to_df_others["L_LS"].append(L_LS)
        dict_to_df_others["ion"].append(particle)
        dict_to_df_others["portion(others)"].append( len(dict_portion["others"])/len(dict_portion["main track"]) )
df_others = pd.DataFrame.from_dict(dict_to_df_others)
df_others.pivot_table(values="portion(others)", columns=["ion"], index=["L_LS"])
```
