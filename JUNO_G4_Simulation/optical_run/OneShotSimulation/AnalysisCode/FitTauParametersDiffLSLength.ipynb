# Get dE/dx Span

```python
dict_dEdx_to_df = {"ion":[], "L_LS":[], "dE/dx":[], "sigma(dE/dx)":[], "N_nuclei":[]}
for L_LS in v_L_LS:
    with np.load(f"/afs/ihep.ac.cn/users/l/luoxj/JUNO_G4_Simulation/optical_run/OneShotSimulation/L_LS_fitResults/dEdx_Span_{L_LS}mm.npz", allow_pickle=True) as f:
        dict_dEdxSpan = f["dict_dEdx_span"].item()
    for particle, v_range in dict_dEdxSpan.items():
        dict_dEdx_to_df["ion"].append(particle)
        dict_dEdx_to_df["L_LS"].append(L_LS)
        dict_dEdx_to_df["dE/dx"].append(np.mean(v_range))
        dict_dEdx_to_df["sigma(dE/dx)"].append(v_range[1]-np.mean(v_range))
        dict_dEdx_to_df["N_nuclei"].append(int(particle.split("_")[1]))
df_dEdx = pd.DataFrame.from_dict(dict_dEdx_to_df).set_index("N_nuclei")
df_dEdx
```

```python
n_figures = 6
colors = ["b","g","r","c","m","y","k","w"]
v_colors = list(sns.color_palette("bright"))[:len(v_L_LS)]
dict_marker = {-1:"o", 0:">", 1:"*"}
dict_marker_color = {-1:"black", 0:"darkorange", 1:"olivedrab"}
fig,axs = plt.subplots(n_figures,1,sharex="col",figsize=(8,9))

for j, L_LS in enumerate(v_L_LS ):
    df_pars_subset_source = df_pars[df_pars["L_LS"]==L_LS]
    for i_constant, name_time_constant in enumerate(v_name_timing_constant):
        v_x = np.array(df_model.index)
        v_y = np.array(df_model[name_time_constant] )

        v_yerr = df_pars_subset_source[name_time_constant+"_error"]
        v_xerr_raw = list(dict_xerr.values())
        if len(v_yerr) == len(v_xerr_raw):
            v_xerr = np.array( list(dict_xerr.values()) )/float(L_LS)
        elif len(v_yerr) == len(v_xerr_raw)-1:
            v_xerr = np.array( list(dict_xerr.values()))[np.array( list(dict_xerr.keys()) )!=10 ]/float(L_LS)
        else:
            print(len(v_y), len(v_xerr_raw))


        axs[i_constant].plot(v_x, v_y, color=colors[i_constant], label=name_time_constant if j==0 else "")
        axs[i_constant].scatter(v_x,v_y, color=colors[i_constant])
    #
        # axs[i_constant].errorbar(df_pars_subset_source["dE/dx"], df_pars_subset_source[name_time_constant],
        #                          yerr=v_yerr,xerr=,
        #                          color=v_colors[j], ls='none', marker="o", capsize=5, capthick=1,
        #                          ecolor=v_colors[j],
        #                          markersize=5, label=f"{L_LS} mm"  if i_constant==j else "")

        # display(df_dEdx[df_dEdx["L_LS"]==L_LS]["dE/dx"])
        axs[i_constant].errorbar(df_dEdx[df_dEdx["L_LS"]==L_LS].loc[df_pars_subset_source["N_nuclei"]]["dE/dx"], df_pars_subset_source[name_time_constant],
                                 yerr=v_yerr,xerr=df_dEdx[df_dEdx["L_LS"]==L_LS].loc[df_pars_subset_source["N_nuclei"]]["sigma(dE/dx)"],
                                 color=v_colors[j], ls='none', marker="o", capsize=5, capthick=1,
                                 ecolor=v_colors[j],
                                 markersize=5, label=f"{L_LS} mm"  if i_constant==j else "")
        axs[i_constant].set_xlim(0,65)
        axs[i_constant].legend()
    plt.xlabel("dE/dx [ MeV/mm ]")
    axs[0].set_title("Time Constant")
```
