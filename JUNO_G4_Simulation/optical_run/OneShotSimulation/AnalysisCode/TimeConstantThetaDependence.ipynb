# Load Time Constant Model

```python
v_name_timing_constant = ["N1", "tau1", "N2", "tau2", "N3", "tau3"]
dir_v_timing_constant = {}

dir_mean_dE_dx_with_quench = {'alpha': 122.773224, 'Co60': 0.6747912, 'AmC': 29.574236}
dir_v_timing_constant[dir_mean_dE_dx_with_quench["Co60"]] = [79.9,4.93,17.1,20.6,3,190]
dir_v_timing_constant[dir_mean_dE_dx_with_quench["AmC"]] = [65,4.93,23.1,34,11.9,220]
dir_v_timing_constant[dir_mean_dE_dx_with_quench["alpha"]] = [65,4.93,22.8,35,12.2,220]

for particle in dir_v_timing_constant.keys():
    for i in range(len(v_name_timing_constant)):
        if "N" in v_name_timing_constant[i]:
            dir_v_timing_constant[particle][i] = dir_v_timing_constant[particle][i]*0.01
df_model = pd.DataFrame.from_dict(dir_v_timing_constant, orient="index", columns=v_name_timing_constant)
df_model
```

# Interpolate Time Constant Input

```python
from scipy.interpolate import interp1d
v_dEdx_to_get = [0.34,11.95, 33.67 ]
v_ion_to_get = ["H_2", "C_12", "Ne_20"]
dict_func_model = {}
df_model_to_plot = pd.DataFrame()
for i, dEdx in enumerate(v_dEdx_to_get):
    dict_tmp = {}
    for time_constant in df_model.columns:
        f = interp1d(np.array(df_model.index), np.array(df_model[time_constant]), fill_value="extrapolate")
        dict_tmp[time_constant] = f(dEdx)
    df_model_to_plot = pd.concat((df_model_to_plot, pd.DataFrame(dict_tmp, index=[v_ion_to_get[i]]) ))

df_model_to_plot.reindex(sorted(df_model_to_plot.columns), axis=1)
```

```python
n_figures = 6

v_marker = ["o", "v", "X", "*"]
fig,axes = plt.subplots(n_figures,1,sharex="col",figsize=(8,9), tight_layout=True)
colors = ["b","g","r","c","m","y","k","w"]
dict_replace = {2:"H_2", 12:"C_12", 20:"Ne_20"}
v_constant_order = [f"tau{i}" for i in range(1,4)]+[f"N{i}" for i in range(1,4)]
for i, N in enumerate( set(df_pars["N_nuclei"]) ):
    # df_pars_to_plot = df_pars[df_pars["N_nuclei"]==N]
    # set error bar
    df_pars_error = pd.DataFrame()
    df_pars_to_plot = pd.DataFrame()
    df_pars_to_plot["theta"] = df_pars["theta"]
    for key in [f"tau{i}_error" for i in range(1,4)]+[f"N{i}_error" for i in range(1,4)]:
        df_pars_to_plot[key.split("_")[0]+f"({dict_replace[N]})"] = df_pars[df_pars["N_nuclei"]==N][key.split("_")[0]]
        df_pars_error[key.split("_")[0]+f"({dict_replace[N]})"] = df_pars[df_pars["N_nuclei"]==N][key]
        # df_pars_error[key.split("_")[0]+f"({dict_replace[N]})"] =

    # plotting
    df_pars_to_plot[["theta"]+[f"tau{i}"+f"({dict_replace[N]})" for i in range(1,4)]+[f"N{i}"+f"({dict_replace[N]})" for i in range(1,4)]].plot(x="theta",ax=axes,
                                                                                        subplots=True, rot=60,
                                                                                          yerr=df_pars_error,
                                                                                          xerr=2.06,
                                                                                          ls='none', marker=v_marker[i], capsize=5, capthick=1,
                                                                                            markersize=5,
                                                                                            color=colors[i])

    for j, ax in enumerate(axes):
        ax.axhline( y=df_model_to_plot.loc[dict_replace[N]][v_constant_order[j]],color=colors[i], linestyle="--",
                    label="model input" if ( (j==5) and (i ==2 )) else None)
        ax.legend(bbox_to_anchor=(1.0, 1.0))
axes[0].set_title("Time Constants (Theta Dependence)")


```
