```python
from scipy.stats import poisson
mu = 0.02
# x = np.arange(poisson.ppf(0.01, mu),
#               poisson.ppf(0.99, mu))
x = range(0,3)
rv = poisson(mu)
plt.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
        label='frozen pmf')
plt.legend(loc='best', frameon=False)
np.random.poisson(0.02)
print(rv.pmf(x))
```
