```python
import random
plt.rcParams["axes.unicode_minus"] = False
sns.set_style("dark")
sns.set_context("paper",font_scale=1.4)
v_x = [random.gauss(0,10) for _ in range(1000)]
v_y = [random.gauss(1,1) for _ in range(1000)]

sns.jointplot(v_x, v_y,kind="hist",bins=20)
plt.xlabel("X")
```

```python
tips_df = sns.load_dataset("tips")
sns.pairplot(tips_df, hue="sex", palette="Blues")
tips_df.head()
```

```python
sns.barplot(x="sex", y="total_bill", data=tips_df)
```

```python
sns.countplot(y="sex", data=tips_df)
```

```python
sns.boxplot(x="day",y="total_bill", data=tips_df, hue="sex")
```

```python
sns.stripplot(x="day",y="total_bill", data=tips_df, hue="sex",jitter=True, dodge=True)
```

```python
sns.swarmplot(x="day",y="total_bill", data=tips_df, hue="sex")
```

```python
tips_mx = tips_df.corr()
sns.heatmap(tips_mx,annot=True)
```
