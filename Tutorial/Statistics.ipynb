```python
from scipy.stats import poisson
v_x_to_df = []
v_y_to_df = []
v_mu_to_df = []
for mu in [0.05,0.1,0.15,0.2,0.3 ]:
    # plt.figure()
    x = np.arange(poisson.ppf(0.01, mu), poisson.ppf(0.99,mu)+1)
    v_x_to_df += list(x)
    v_y_to_df += list(poisson.pmf(x, mu)) 
    v_mu_to_df += [mu]*len(x)
    # plt.xlim(-0.1,4)
    # plt.xlabel("NPE")
    # plt.ylabel("Probability")
    # plt.title(f"$\mu$={mu}")
df = pd.DataFrame(zip(v_x_to_df, v_mu_to_df, v_y_to_df), columns=["n", "mu", "Probability"])
sns.barplot(data=df, x="n", y="Probability", hue="mu")
plt.title("Poisson Distribution")
    
```
