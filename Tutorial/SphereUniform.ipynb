```python
import matplotlib.pylab as plt
import numpy as np
plt.style.use(["science","no-latex", "/afs/ihep.ac.cn/users/l/luoxj/Style/Paper.mplstyle"])
%config InlineBackend.figure_format = 'svg'
#plt.rcParams['figure.dpi'] = 300
#plt.rcParams['savefig.dpi'] = 300
import sys
sys.path.append("/afs/ihep.ac.cn/users/l/luoxj/root_tool/python_script/")
```

```python
from random import random
v_x = []
v_y = []
v_z = []
for i in range(10000):
    u = random()
    v = random()
    theta = 2 * np.pi * u
    phi = np.arccos(2*v-1)
    v_x.append( np.sin(theta)*np.cos(phi) )
    v_y.append( np.sin(theta)*np.sin(phi) )
    v_z.append( np.cos (theta) )
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.scatter(v_x, v_y, v_z, s=3)
```

```python
plt.hist(v_z, bins=100,histtype="step", label="z")
plt.hist(v_x, bins=100,histtype="step", label="x")
plt.hist(v_y, bins=100,histtype="step", label="y")
plt.legend()
```
